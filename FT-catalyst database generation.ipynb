```python
ndb.write_superstructure_db_to_brightway(name="Superstructure db FT-catalyst")
```

The following manual change was made to the database:
A copy was made of the activity "selective catalytic reduction of nitogen oxides, GLO". The copy was renamed and a biosphere exchange of 0.11 kg NOx to the air - non-urban air or from high stacks was added. This accounts for a 90% NOx removal efficiency of current DeNOx units. The changed activity is used in the foreground model of the Fischer-Tropsch catalyst production process
